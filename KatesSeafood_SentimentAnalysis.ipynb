## Load and Clean Data

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = pd.read_excel("KatesReviews.xlsx")
data
```

```python
data.shape
```

```python
data2 = data.copy()
data2
```

```python
def get_city(loc):
    list_loc = loc.split(", ")
    city = pd.Series(list_loc[0])
    return city

def get_state(loc):
    list_loc = loc.split(", ")
    state = pd.Series(list_loc[1])
    return state
```

```python
data2["reviewer_city"] = data2["reviewer_loc"].apply(lambda loc: get_city(loc))
data2["reviewer_state"] = data2["reviewer_loc"].apply(lambda loc: get_state(loc))
data2
```

```python
def convert_score(score):
    num_score = int(score.split(" ")[0])
    return num_score
```

```python
data2["rating"] = data2["star_rating"].apply(lambda score: convert_score(score))
data2
```

```python
# drop review number column
data2.drop('Unnamed: 0', axis=1, inplace=True)
data2
```

```python
# look for missing values in dataset
data2.isna().sum()
```

## Sentiment Analysis

```python
import re
```

```python
def clean_reviews(sentence, regex):

    no_punc = regex.sub('', sentence)
    lowercase = no_punc.lower()
    split_sent = lowercase.split(' ')
    return split_sent
```

```python
# Create column for list of words in each review

regex = re.compile('[^a-zA-Z,  ]')
data2['review_bag'] = data2['review_text'].apply(lambda sent: clean_reviews(sent, regex))
data2
```

```python
# Open lists of positive and negative words
positives = set(open('positive.txt').read().split())
negatives = set(open('negative.txt').read().split())

# Get counts of positive and negative words in the reviews
data2["positives"] = data2['review_bag'].apply(lambda row: len([word for word in row if word in positives]))
data2["negatives"] = data2['review_bag'].apply(lambda row: len([word for word in row if word in negatives]))

# Get review word count
data2["word_count"] = data2['review_bag'].apply(lambda row: len(row))
data2
```

```python
data2['pos_neg_diff'] = data2['positives'] - data2['negatives']
data2
```
